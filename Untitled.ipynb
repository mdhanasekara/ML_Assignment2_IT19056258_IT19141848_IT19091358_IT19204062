```python
yes_no_mapping = {'Yes' : 1, 'No' :0, 'No Record':0}
customer_travel_churn_data['FrequentFlyer']=customer_travel_churn_data['FrequentFlyer'].map(yes_no_mapping)
customer_travel_churn_data['AccountSyncedToSocialMedia']=customer_travel_churn_data['AccountSyncedToSocialMedia'].map(yes_no_mapping)
customer_travel_churn_data['BookedHotelOrNot']=customer_travel_churn_data['BookedHotelOrNot'].map(yes_no_mapping)
```

```python
customer_travel_churn_data
```

```python
AnnualIncome_mapping = {'Middle Income' : 'M', 'Low Income' :'L', 'High Income':'H'}
customer_travel_churn_data['AnnualIncomeClass']=customer_travel_churn_data['AnnualIncomeClass'].map(AnnualIncome_mapping)
```
